# 01 — Exploratory Data Analysis
Project 08: Topic Modeling with 20 Newsgroups

This notebook explores the raw and preprocessed corpus before any modeling.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

from src.data.make_dataset import load_processed, build_dataset
from src.utils import load_config, ensure_dirs
from src.visualization.plots import (
    plot_doc_length_distribution,
    plot_category_distribution,
    plot_top_words_global,
    save_all_eda_plots,
)

ensure_dirs()
cfg = load_config()
%matplotlib inline
```

## 1. Load Dataset

```python
try:
    df = load_processed()
except FileNotFoundError:
    print('Building dataset for the first time...')
    df = build_dataset(cfg)

print(f'Documents: {len(df):,}')
print(f'Categories: {df["category_name"].nunique()}')
df.head(3)
```

## 2. Category Distribution

```python
fig = plot_category_distribution(
    df['category_name'].tolist(),
    save_path='../reports/figures/category_distribution.png'
)
plt.show()
```

## 3. Document Length Distribution

```python
df['raw_len']   = df['raw_text'].str.split().str.len()
df['clean_len'] = df['clean_text'].str.split().str.len()

print('--- Raw text ---')
print(df['raw_len'].describe().round(1))
print('\n--- Clean text ---')
print(df['clean_len'].describe().round(1))

fig = plot_doc_length_distribution(
    df['raw_text'].tolist(),
    save_path='../reports/figures/raw_doc_lengths.png'
)
plt.show()

fig = plot_doc_length_distribution(
    df['clean_text'].tolist(),
    title='Cleaned Document Length Distribution',
    save_path='../reports/figures/clean_doc_lengths.png'
)
plt.show()
```

## 4. Top Words (after preprocessing)

```python
fig = plot_top_words_global(
    df['clean_text'].tolist(),
    n=30,
    save_path='../reports/figures/top_words_global.png'
)
plt.show()
```

## 5. Per-category Word Frequency Snapshot

```python
sample_cats = df['category_name'].unique()[:4]

fig, axes = plt.subplots(2, 2, figsize=(14, 8))
for ax, cat in zip(axes.flatten(), sample_cats):
    texts = df[df['category_name'] == cat]['clean_text']
    tokens = ' '.join(texts).split()
    common = Counter(tokens).most_common(15)
    words, counts = zip(*common)
    ax.barh(words[::-1], counts[::-1], color='#2E86AB')
    ax.set_title(cat, fontsize=9, fontweight='bold')
    ax.tick_params(labelsize=8)

fig.suptitle('Top 15 Words by Category (sample)', fontsize=12)
fig.tight_layout()
fig.savefig('../reports/figures/per_category_words.png', dpi=150, bbox_inches='tight')
plt.show()
```
