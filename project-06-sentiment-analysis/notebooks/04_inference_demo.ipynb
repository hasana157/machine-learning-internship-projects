```python
import joblib
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

# Ensure required NLTK data is available
nltk.download('stopwords')
nltk.download('punkt')

print("Imports and downloads complete.")
```

```python
# Load the pipeline (TF‑IDF + Logistic Regression)
model_path = '../models/sentiment_pipeline.joblib'
pipeline = joblib.load(model_path)
print("Model loaded successfully.")
```

```python
def clean_text(text, remove_stopwords=True, use_stemming=True):
    """
    Clean and preprocess text data (same as used during training).
    """
    # Lowercase
    text = text.lower()
    
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    
    # Remove punctuation and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Tokenize
    words = text.split()
    
    # Remove stopwords
    if remove_stopwords:
        stop_words = set(stopwords.words('english'))
        words = [w for w in words if w not in stop_words]
    
    # Stemming
    if use_stemming:
        stemmer = PorterStemmer()
        words = [stemmer.stem(w) for w in words]
    
    return ' '.join(words)
```

```python
def predict_sentiment(text, model, show_prob=False):
    """
    Clean the input text, then predict sentiment.
    Returns: 'Positive' or 'Negative', and optionally probabilities.
    """
    # Clean the text
    cleaned = clean_text(text)
    
    # Predict
    pred = model.predict([cleaned])[0]
    sentiment = 'Positive' if pred == 1 else 'Negative'
    
    if show_prob:
        proba = model.predict_proba([cleaned])[0]
        return sentiment, proba
    return sentiment
```

```python
# Example positive reviews
test_positive = [
    "This movie was absolutely fantastic! I loved every moment.",
    "The acting was superb and the story kept me engaged.",
    "A masterpiece – definitely worth watching again."
]

# Example negative reviews
test_negative = [
    "Terrible film, complete waste of time.",
    "The plot was boring and the acting was wooden.",
    "I couldn't even finish it. Disappointing."
]

print("Positive predictions:")
for text in test_positive:
    result = predict_sentiment(text, pipeline, show_prob=True)
    print(f"Text: {text[:50]}... -> Sentiment: {result[0]}  (Probabilities: Neg={result[1][0]:.3f}, Pos={result[1][1]:.3f})")

print("\nNegative predictions:")
for text in test_negative:
    result = predict_sentiment(text, pipeline, show_prob=True)
    print(f"Text: {text[:50]}... -> Sentiment: {result[0]}  (Probabilities: Neg={result[1][0]:.3f}, Pos={result[1][1]:.3f})")
```

```python
print("Enter a review (or 'quit' to exit):")
while True:
    user_input = input("> ")
    if user_input.lower() == 'quit':
        break
    if user_input.strip() == '':
        continue
    sentiment = predict_sentiment(user_input, pipeline)
    print(f"Predicted sentiment: {sentiment}")
```
